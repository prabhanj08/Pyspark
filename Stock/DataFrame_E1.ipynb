```
from pyspark import SQLContext
from pyspark.sql import SparkSession
```

```
sparksc = SparkSession.builder.appName('StockEx').getOrCreate()
```

```
stock_df = spark.read.csv('/FileStore/tables/appl_stock.csv',header=True,inferSchema=True)
stock_df.show(3,False)
```

```
#Filtering Operations

#Filter records having close price less than 500

## Pass on a SQL syntax to the filter() method
stock_df.filter('close<500').show(5,False)
```

```
# Filter Specific column
stock_df.filter('close<500').select('Date','Open','High','low','close').show(3,False)
```

```
stock_df.filter('close < 500 and open > 500' ).show(3,False)
```

```
## Using Dataframes

stock_df.filter(stock_df['close'] < 500).show(3,False)
```
